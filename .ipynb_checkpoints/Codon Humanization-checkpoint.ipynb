# DNA2vec

```python
file_name = 'dna2vec-20161219-0153-k3to8-100d-10c-29320Mbp-sliding-Xat.w2v'
k = 3
```

```python
file_path = './data/' + file_name
f = open()
lines = f.readlines()
f.close()
```

```python
import numpy as np
w2v = dict()
# w2v = word to 100-dimensional vector
for line in lines[1:]:
    s = line.split()
    key = s[0]
    value = s[1:]
    value = np.array(list(map(float, value)))
    w2v[key] = value
```

# Dataset configuration

```python
# Human sequences
from Bio import SeqIO
from tqdm import tqdm

seqs = SeqIO.parse('./data/GRCh38_latest_rna.fna', "fasta")

human_seqs = []
for seq in tqdm(seqs):
    human_seqs.append(str(seq.seq) + '\n')
human_seqs = human_seqs[::5]
with open('./human_seqs.txt', 'w') as f:
    f.writelines(human_seqs)
```

```python
# Random sequences
import random
random.seed(42)
base = ['A', 'T', 'G', 'C']

def make_random_seq(seq):
    new_seq = []
    for _ in range(len(seq.strip())):
        new_seq.append(random.choice(base))
    return ''.join(new_seq)

random_seqs = []
for seq in tqdm(human_seqs):
    random_seqs.append(make_random_seq(seq))
```

```python
# Calculation CAI for human and random sequences
from Bio.SeqUtils import CodonUsage as CU
from Bio.Seq import Seq

human_cai = []
random_cai = []

for seq in tqdm(human_seqs):
    try:
        seq = seq.strip()
        seq = seq[:len(seq)-len(seq)%3]
        seq = Seq(seq)
        myIndex = CU.CodonAdaptationIndex()
        v = myIndex.cai_for_gene(seq)
        human_cai.append(v)
    except:
        pass
    
for seq in tqdm(random_seqs):
    try:
        seq = seq.strip()
        seq = seq[:len(seq)-len(seq)%3]
        seq = Seq(seq)
        myIndex = CU.CodonAdaptationIndex()
        v = myIndex.cai_for_gene(seq)
        random_cai.append(v)
    except:
        pass
```

# Sequence vectorization and concat positive and negatie dataset

```python
def make_kmer(seq, k):
    n = len(seq)
    bags = []
    for i in range(len(seq)):
        if i+k>n: break
        bags.append(seq[i:i+k])
    return bags

def vectorizer(seq, k=3):
    seq = seq.strip()
    kmers = make_kmer(seq, k)
    v = np.zeros(100)
    for kmer in kmers:
        v += w2v[kmer]
    return v
```

```python
X = []
y = []
k = 3

for seq in tqdm(human_seqs):
    try:
        v = vectorizer(seq, k)
        X.append(v)
        y.append(1)
    except:
        pass
    
for seq in tqdm(random_seqs):
    try:
        v = vectorizer(seq, k)
        X.append(v)
        y.append(0)
    except:
        pass
```

# Define silent mutation

```python
import re
import random

codon_table = {
    'F':['TTT', 'TTC'],
    'L':['TTA', 'TTG', 'CTT', 'CTC', 'CTA', 'CTG'],
    'I':['ATT', 'ATC', 'ATA'],
    'M':['ATG'],
    'V':['GTT', 'GTC', 'GTA', 'GTG'],
    'S':['TCT', 'TCC', 'TCA', 'TCG'],
    'P':['CCT', 'CCC', 'CCA', 'CCG'],
    'T':['ACT', 'ACC', 'ACA', 'ACG'],
    'A':['GCT', 'GCC', 'GCA', 'GCG'],
    'Y':['TAT', 'TAC'],
    '*':['TAA', 'TAG', 'TGA'],
    'H':['CAT', 'CAC'],
    'Q':['CAA', 'CAG'],
    'N':['AAT', 'AAC'],
    'K':['AAA', 'AAG'],
    'D':['GAT', 'GAC'],
    'E':['GAA', 'GAG'],
    'C':['TGT', 'TGC'],
    'W':['TGG'],
    'R':['CGT', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG'],
    'S':['AGT', 'AGC'],
    'G':['GGT', 'GGC', 'GGA', 'GGG']
}

def make_point_mutation(seq):
    codons = re.findall('\w\w\w', seq)
    c = random.randint(0, len(codons)-1) # make point mutation on n_th codon
    codon_list = codon_table[str(Seq(codons[c]).translate())]
    codons[c] = random.choice(codon_list) # point mutation
    return ''.join(codons)
```

# Genetic algorithm for codon humanization

```python
# sequence optimization for given sequence
original_seq = make_random_seq("A"*3*300) #150
print("Original Sequence: ", original_seq)
pool = []
n = 100 # n generation
k = 100 # k seqeunces in pool
a = 10 # 10 try per sequence in a generation
early_stopper = 0.9999
probs = []

# round 0, initialization
pool.append(list())
for _ in range(k):
    pool[-1].append(original_seq)

probs.append(xgmodel.predict_proba([vectorizer(original_seq)])[0][1])
print("[Round 0] ", probs[0])
for r in range(n):
    new_pool = list()
    new_pool_probs = []
    print(f'[Round {r+1}]')
    for seq in pool[-1]:
        original = xgmodel.predict_proba([vectorizer(seq)])[0][1]
        for _ in range(a):
            new_seq = make_point_mutation(seq)
            improved = xgmodel.predict_proba([vectorizer(new_seq)])[0][1]
            if original < improved:
                new_pool.append(new_seq)
                new_pool_probs.append(improved)
    result_list = [i for _,i in sorted(zip(new_pool_probs,new_pool), reverse=True)]
    probs.append(sum(new_pool_probs)/len(new_pool_probs))
    print(f'result: {probs[-1]}')
    pool.append(new_pool[:k])
    if probs[-1] > early_stopper:
        print("Early Stopping..")
        break
```

```python
# Calculation average CAI for each generation
codon_index = []
for p in pool:
    indexs = []
    for seq in p:
        myIndex = CU.CodonAdaptationIndex()
        v = myIndex.cai_for_gene(seq)
        indexs.append(v)
    codon_index.append(sum(indexs)/len(indexs))
```

```python
# Draw figure 7
plt.figure(dpi=400)
plt.plot(range(len(probs)), probs, label='Humanity')
plt.xlabel('Round')
#plt.ylabel('Humanity')

plt.plot(range(len(codon_index)), codon_index, label='CAI')
plt.xlabel('Generation')
#plt.ylabel('Codon Adaptation Index(CAI)')

plt.legend()
```
